```python
u = read_train.joy_x.value_counts().sort_index()
plt.figure(figsize=(5, 2))
plt.rcParams.update({'font.size': 10}) # Set larger plot font size
plt.bar(u.index, u.values)
plt.xlabel('Joy_x')
plt.ylabel('counts')
plt.show()
```

```python
y = read_train.joy_y.value_counts().sort_index()
plt.figure(figsize=(5, 2))
plt.rcParams.update({'font.size': 10}) # Set larger plot font size
plt.bar(y.index, y.values)
plt.xlabel('Joy_y')
plt.ylabel('counts')
plt.show()
```
