```python
# converted emotion  score to their respective emotion labels such as joy, sadness, anticipation, 
# disgust etc as annoted by the annotators
```

```python
train_data = pd.read_excel('dataset/train2.xlsx')
test_data = pd.read_excel('dataset/test3.xlsx')
```

```python
train_data.head()
```

```python
test_data.head()
```

```python
print(f'Number of trained data: {len(train_data)}\nNumber of test data: {len(test_data)}')
```

```python
#Our models will be train and test based on the two columns that is text and emotion columns. 
# The text column will be converted into vector format
```

```python

X_train = train_data['text']
y_train = train_data['emotion_x']
X_test = test_data['text']
y_test = test_data['emotion_x']
```

```python
pipeline = Pipeline([
    ('vectorizer', TfidfVectorizer()),
    ('classifier', LinearSVC())
])
```

```python
pipeline.fit(X_train, y_train)
```

```python
#training accuracy
train_accuracy = pipeline.score(X_train, y_train)
print('Accuracy:', train_accuracy)
```

```python
test_accuracy = pipeline.score(X_test, y_test)
print('Accuracy:', test_accuracy)
```

```python
# the second method which SVC will also be train using thesame column from the dataset
```

```python
from sklearn.svm import SVC
```

```python
vectorizer = TfidfVectorizer()
```

```python
X = vectorizer.fit_transform(X_train)
Y = vectorizer.fit_transform(X_test)
```

```python
clf = SVC(kernel='linear')
clf.fit(X, y_train)
```

```python
#train accuracy
train_accuracy = clf.score(X, y_train)
print("Accuracy:", train_accuracy)
```

```python
#method3 which is the naive bayes will also be trained using thesame column from the dataset
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
vectorizer = CountVectorizer()
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

```

```python
nb = MultinomialNB()
nb.fit(X_train_vec, y_train)
```

```python
#train accuracy
train_accuracy = nb.score(X_train_vec, y_train)
print("Accuracy: {:.2f}".format(train_accuracy))
```

```python
#test accuracy
test_accuracy = nb.score(X_test_vec, y_test)
print("Accuracy: {:.2f}".format(test_accuracy))
```
